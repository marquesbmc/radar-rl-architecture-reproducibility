# Article Results Analysis

This notebook analyzes the eighteen simulations used in the manuscript. Supplementary PER experiments are intentionally excluded from the main table.

```python
from pathlib import Path
import pandas as pd

ROOT = Path.cwd().resolve().parent if Path.cwd().name == "notebooks" else Path.cwd().resolve()
DATA = ROOT / "results" / "article" / "sim_summary_article.csv"
frame = pd.read_csv(DATA)
frame.head()
```

```python
assert len(frame) == 18
assert "per" not in set(frame["algorithm"])
assert set(frame["scenario"]) == {"standard", "dense_restricted", "extensive_sparse"}
print("Validated: 18 article simulations")
```

## Configuration summary

```python
frame.groupby(["architecture", "algorithm", "scenario"]).size().rename("rows").to_frame()
```

## Mean interaction success

```python
frame.pivot_table(
    index=["algorithm", "scenario"],
    columns="architecture",
    values="combined_mean_successes",
    aggfunc="mean",
)
```

## Collaborative effectiveness

```python
frame.pivot_table(
    index=["algorithm", "scenario"],
    columns="architecture",
    values="combined_mean_collaborative_effectiveness",
    aggfunc="mean",
)
```

## Spatial exploration

```python
frame.pivot_table(
    index=["algorithm", "scenario"],
    columns="architecture",
    values="combined_mean_unique_positions",
    aggfunc="mean",
)
```
